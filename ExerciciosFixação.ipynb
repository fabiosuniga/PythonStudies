```python
#Este exercício nos retorna um dicionário com os bairros e a distância total a ser percorrida em cada um.

pacotes_do_dia = [
    (101, 'Centro', 5.5, 'Pendente'),
    (102, 'Bairro Alto', 12.0, 'Entregue'),
    (103, 'Centro', 2.0, 'Pendente'),
    (104, 'Zona Sul', 8.5, 'Pendente'),
    (105, 'Bairro Alto', 4.0, 'Pendente'),
    (106, 'Zona Sul', 3.2, 'Entregue')
] #lista de tuplas com os dados.

def calcular_distancia_pendente(lista_pacotes):
  pendencias = [ (bairro, distancia) for _, bairro, distancia, status in lista_pacotes if status == 'Pendente' ] # filtragem com List comprehension.
  rotas_pendencias = {}
  for bairro, km in pendencias:
    rotas_pendencias[bairro] = rotas_pendencias.get(bairro, 0) + km # Se não encontrar bairro ele cria um novo, e adiciona o km do item atual do "for".

  return rotas_pendencias

print(calcular_distancia_pendente(pacotes_do_dia))
```

```python
#Este exercício junta pessoas pelo setor que elas trabalham

novos_contratados = [
    ('Ana', 'TI'),
    ('Carlos', 'Vendas'),
    ('Beatriz', 'TI'),
    ('Daniel', 'RH'),
    ('Eduardo', 'Vendas'),
    ('Fernanda', 'TI')
]

def agrupar_departamentos(lista_funcionarios):
    rh_atualizado = {}
    for nome, setor in lista_funcionarios:
        rh_atualizado[setor] = rh_atualizado.get(setor, []) #Se não achar o setor, cria uma lista vazia.
        rh_atualizado[setor].append(nome) # Adiciona o nome do funcionario no setor.

    return rh_atualizado

print(agrupar_departamentos(novos_contratados))
```

```python
# Verificação simples para critérios de segurança de uma senha.

def validar_senha(senha):
  condicoes = [  # uma lista de condições para que a senha seja validada
      len(senha) > 7, # senha maior que 7 letras.
      any(char.isupper() for char in senha), # ao menos uma letra maiúscula.
      any(char.isdigit() for char in senha) # ao menos um número
  ]

  return all(condicoes) #returna true se todas condições foram aceitas.

senha = 'Fabio123'
print(validar_senha(senha))
```

```python
# Exercicio que separa quem comprou ingressos para os dois dias de evento, quem comprou somente sábado e o número total de pessoas.

ingressos_sabado = ['ana@email.com', 'bruno@email.com', 'carlos@email.com', 'ana@email.com', 'edu@email.com']
ingressos_domingo = ['carlos@email.com', 'daniel@email.com', 'eduardo@email.com', 'bruno@email.com', 'carlos@email.com']

def analise_ingressos(lista_sab, lista_dom):
  setsabado = set(lista_sab) #sets removem duplicados.
  setdomingo = set(lista_dom)
  analise_ingresso = { #é possível criar união, interseção e diferença de sets.
      'ambos_dias': list(setsabado.intersection(setdomingo)),
      'somente_sabado': list(setsabado.difference(setdomingo)),
      'total_unicos': len(setsabado.union(setdomingo))
  }

  return analise_ingresso

print(analise_ingressos(ingressos_sabado, ingressos_domingo))

```

```python
# Exercicio para separar dados de logs de erro gerados em uma única string

logs_servidor = [
    "ERROR:2026-05-20:Falha na conexão com o banco de dados",
    "INFO:2026-05-20:Usuário 'admin' fez login",
    "WARNING:2026-05-20:Uso de memória acima de 80%",
    "ERROR:2026-05-20:Tempo de resposta excedido na API",
    "INFO:2026-05-20:Backup diário concluído"
]

def agrupar_logs(lista_logs):
  dict_logs = {}
  for linha in lista_logs:
    erro, _, msg = linha.split(":") # usamos ':' para separar cada setor do erro. Só usamos o tipo do erro, e a mensagem dele.
    dict_logs[erro] = dict_logs.get(erro, []) #Se não achar o erro, cria uma lista vazia.
    dict_logs[erro].append(msg) # Adiciona a mensagem no erro percorrido.
  return dict_logs
print(agrupar_logs(logs_servidor))
```

```python
# Exercicio para tratamento de dados que não deveriam ser inseridos na função

precos_carrinho = ['150.50', '45.00', 'gratis', '99.90', 'N/A', '10.00']

def somar_carrinho(lista_precos):
  total = [] # nova lista para termos os dados em float

  for num in lista_precos:
    try:  #bloco try  para tentar converter o string em float.
      total.append(float(num))
    except ValueError: # se não for possível converter ele apenas pula para o próximo valor
      continue

  return sum(total) # retorna a soma de todos.

print(somar_carrinho(precos_carrinho))
```

```python
# orientação a objetos, cria um personagem que toma dano e se cura

class Personagem:

    def __init__(self, nome, hp_maximo): # método que defini quais atributos ele vai ter
        self.nome = nome
        self.hp_maximo = hp_maximo
        self.hp_atual = hp_maximo

    def receber_dano(self, quantidade): # define a quantia de dano recebido
        try:
          self.hp_atual -= quantidade
          print(f"Você tomou {quantidade} de dano!")
          self.hp_atual = 0 if self.hp_atual < 0 else self.hp_atual # operador ternário o HP não pode ficar negativado
        except (ValueError, TypeError):
          print("Digite um número inteiro!")

    def curar(self, quantidade):
        try:
          self.hp_atual += quantidade
          print(f"Você curou {quantidade} de HP!")
          self.hp_atual = min(self.hp_atual, self.hp_maximo) # mesmo curando, ele não pode ultrapassar o hp máximo
        except (ValueError, TypeError):
          print("Digite um número inteiro!")

    def status(self):
        print(f"Nome: {self.nome}\n Hp Máximo: {self.hp_maximo}\n HP atual: {self.hp_atual}")

# --- ÁREA DE TESTES ---

heroi = Personagem("Bram", 100)
heroi.status()
heroi.receber_dano(30)
heroi.status()
heroi.curar(50)
heroi.status()
heroi.receber_dano(200)
heroi.status()
```
